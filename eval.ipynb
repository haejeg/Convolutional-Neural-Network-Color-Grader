```python
# Imports & Data Loading
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

metrics = pd.read_csv('/experiments/run_20260424_000739/results/metrics.csv')
```

```python
metrics.columns
```

```python
# Analysis of epoch vs validation loss
X = metrics[['epoch']].astype(float)
y = metrics['val_loss'].astype(float)

plt.plot(X, y)
plt.xlabel('Epoch')
plt.ylabel('Validation Loss')
# Looks like our validation loss has reached stability around epoch 85. The data doesn't look like it's overfitting either,
# as validation loss isn't rising as epoch increases.
```

```python
# Training loss vs validation loss
y = metrics['train_loss'].astype(float)

plt.plot(X, y)
plt.xlabel('Epoch')
plt.ylabel('Training Loss')
# Training loss is higher than validation loss, which is expected- since we apply transformations to the training data
# that aren't applied to the validation data.
# There's a steady decline in our training loss as epoch increases, which is a good sign.
# There are zigzags, so maybe we should implement a Learning Rate scheduler, or check batch size.
```
